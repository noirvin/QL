```python
def quiz(dataset, lower, upper):
    
    count_upper=0
    count_lower=0
    
    for data in dataset:
        if data >=lower:
            count_lower+=1
        if data <=upper:
            count_upper+=1
    print("There are "+str(count_lower)+" elements that are equal or greater than lower threshold") 
    print("There are "+str(count_upper)+" elements that are equal or less than upper threshold")
            
    
    
```

```python
ls = [1,2,3,4]
test = quiz(ls,2,3)
print(test)
```
