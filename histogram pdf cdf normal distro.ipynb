```python
def quiz(dataset, lower, upper):
    
    count = 0
    
    for data in dataset:
        if data >=lower and data<=upper:
            count+=1
    
    return count
```

```python
ls = [1,2,3,4]
test = quiz(ls,2,3)
print(test)
```

```python
def lower_than_threshold(dataset,threshold):
    
    count = 0
    
    for data in dataset:
        if data<=threshold:
            count+=1   
            
    return count/len(dataset)

```

```python
def cdf_list(dataset, step):
    threshold_range = list(range(int(min(dataset)),int(max(dataset)),step))
    cdf = []
    for threshold in threshold_range:
        
        cdf.append(lower_than_threshold(dataset,threshold))
        
    return cdf    
        
        

```

```python
test_list = cdf_list(ls_age,1)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(int(np.min(ls_age)), int(np.max(ls_age))), test_list)
plt.grid()
```

```python
def gaussian_f(x, mu, sigma):
    f = 1/np.sqrt(2*np.pi*sigma**2)*np.exp(-(x-mu)**2/2/(sigma**2))
    return f

x = np.linspace(-5, 5, 100)

y = [gaussian_f(i, 0, 1)for i in x]

plt.plot(x, y)
plt.grid()
```
